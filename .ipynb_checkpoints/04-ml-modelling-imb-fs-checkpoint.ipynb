# ML Modelling

```python
target = data['Lithology']
```

## Scale the Data

```python
# Feature Scaling
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# scaler = RobustScaler()
# scaled_features = scaler.fit_transform(features)
```

```python
X, y = scaled_features, target.to_numpy()
```

## Class Imbalance Fixes

```python
imb_fixes = {
    'SMOTE': SMOTE(random_state=2024),
    # 'ADASYN': ADASYN(sampling_strategy='minority'),
    'SMOTETomek': SMOTETomek(random_state=2024, sampling_strategy='minority'),
    # 'SMOTEnn': SMOTEENN(),
}
```

```python
# Apply each of the Class Imbalance fix to the features and save as dict object
imb_dict = {}
for imb_name, imb_fix in imb_fixes.items():
    imb = {}

    imb['X'], imb['y'] = imb_fix.fit_resample(X, y)

    imb_dict[imb_name] = imb
```

```python
print(imb_dict.keys())
print(imb_dict['SMOTE'].keys())
print(imb_dict['SMOTE']['y'])
```

## Feature Selection

```python
# Run 5 different Feature selection methods on each of the output of the class imbalance fix
# 10 different model input combinations
fs_dict = {}
fs_cols_dict = {}

for imb_name, imb_data in imb_dict.items():

    fsd = {}
    fscols = {}

    X_imb, y_imb = imb_data['X'], imb_data['y']

    # PCA
    pca = PCA(n_components=0.95)
    X_pca = pca.fit_transform(X_imb)
    Xtn, Xt, ytn, yt = train_test_split(X_pca, y_imb, test_size=0.25, random_state=42)
    fsd['pca'] = {'X_train' : Xtn, 'X_test' : Xt, 'y_train' : ytn, 'y_test' : yt}
    fscols['pca'] = pca

    # KBest Features
    kbest = SelectKBest(score_func=f_classif, k=50)
    X_kbest = kbest.fit_transform(X_imb, y_imb)
    Xtn, Xt, ytn, yt = train_test_split(X_kbest, y_imb, test_size=0.25, random_state=42)
    fsd['kbest'] = {'X_train' : Xtn, 'X_test' : Xt, 'y_train' : ytn, 'y_test' : yt}
    fscols['kbest'] = kbest

    # RFE Features
    rfe = RFE(estimator=DecisionTreeClassifier(random_state=2024), n_features_to_select=50)
    X_rfe = rfe.fit_transform(X_imb, y_imb)
    Xtn, Xt, ytn, yt = train_test_split(X_rfe, y_imb, test_size=0.25, random_state=42)
    fsd['rfe'] = {'X_train' : Xtn, 'X_test' : Xt, 'y_train' : ytn, 'y_test' : yt}
    fscols['rfe'] = rfe

    # Mutual Info Scores
    mi_score = mutual_info_classif(X_imb, y_imb)
    mi_df = pd.DataFrame({'feature':features.columns, 'importance':mi_score})
    mi_df = mi_df.sort_values(by=['importance'], ascending=False).reset_index(drop=True)
    mi_cols = mi_df.head(50).feature.tolist()
    index = [features.columns.get_loc(c) for c in mi_cols if c in features]
    X_mi = X_imb[:, index]
    Xtn, Xt, ytn, yt = train_test_split(X_mi, y_imb, test_size=0.25, random_state=42)
    fsd['mi'] = {'X_train' : Xtn, 'X_test' : Xt, 'y_train' : ytn, 'y_test' : yt}
    fscols['mi_score'] = mi_score

    # RFC Importance
    rfc = RandomForestClassifier(random_state=2024, n_estimators=100)
    xrfc = rfc.fit(X_imb, y_imb)
    rfc_df = pd.DataFrame({'feature':features.columns, 'importance':rfc.feature_importances_})
    rfc_df = rfc_df.sort_values(by=['importance'], ascending=False).reset_index(drop=True)
    rfc_cols = rfc_df.head(50).feature.tolist()
    index = [features.columns.get_loc(c) for c in rfc_cols if c in features]
    X_rfc = X_imb[:, index]
    Xtn, Xt, ytn, yt = train_test_split(X_rfc, y_imb, test_size=0.25, random_state=42)
    fsd['rfc'] = {'X_train' : Xtn, 'X_test' : Xt, 'y_train' : ytn, 'y_test' : yt}
    fscols['rfc'] = rfc

    fs_dict[imb_name] = fsd
    fs_cols_dict[imb_name] = fscols
```

# Run Models

```python
models = {
    # 'LightGBM' : LGBMClassifier(),
    'Logit' : LogisticRegression(random_state=2024, class_weight='balanced'),
    'DTree' : DecisionTreeClassifier(random_state=2024, class_weight='balanced'),
    'RForest' : RandomForestClassifier(random_state=2024, class_weight='balanced'),
    'AdaBoost' : AdaBoostClassifier(random_state=2024),
    'GBoost' : GradientBoostingClassifier(random_state=2024),
}
```

```python
# Run 5 ML Models on 10 different input types based on Class Imbalance fix and Feature selection methods
# 50 different model types

metrix = dict()
model_dict = dict()
# cv_scores = dict()

for model_name, model in tqdm(models.items(), desc='Looping ML Models'):

    for imb_name, fs_data in fs_dict.items():

        for fs_name, dt in fs_data.items():

            # print(f'{model_name} | {imb_name} | {fs_name}')
            key = f'{model_name}-{imb_name}-{fs_name}'

            X_train, X_test, y_train, y_test = dt['X_train'], dt['X_test'], dt['y_train'], dt['y_test']

            model.fit(X_train, y_train) 
            y_pred = model.predict(X_test)
            y_probs = model.predict_proba(X_test)[:, 1]

            metrix[key] = classification_report(y_test, y_pred, zero_division=0, digits=3)
            model_dict[key] = model 
            # cv_scores[key] = cross_val_score(model, X_train, y_train, cv=3)
```

```python
for k, v in metrix.items():
    print(k)
    print(v)
    print('---')
```

#### **What's Working Well**:

* The models that perform best tend to have a good balance between precision and recall across different classes, as evidenced by their F1-scores.
* Models like Random Forest and Gradient Boosting are effective in handling complex data patterns, which is a common characteristic of datasets with multiple classes.
* Feature selection methods (kbest, rfe) combined with SMOTE (Synthetic Minority Over-sampling Technique) show promising results, indicating that both feature selection and addressing class imbalance are critical for model performance.

#### **Areas for Improvement**:

* While some models achieve decent accuracy, there is still room for improvement in precision and recall for certain classes, as indicated by lower F1-scores in some cases.
* Models like Decision Tree and Logistic Regression, even with SMOTE and feature selection techniques, do not consistently reach the accuracy levels of Random Forest and Gradient Boosting.
* The variation in performance across different models and techniques suggests that the dataset might be complex with features that require careful preprocessing and feature engineering.

#### **Data Complexity**:

* The variation in model performance indicates the complexity of the data. Certain classes are more challenging to predict accurately, possibly due to fewer samples or more complex patterns within those classes.
* The success of ensemble methods like Random Forest and Gradient Boosting suggests that the dataset benefits from models that can capture complex interactions between features.

#### **Suggestions for Further Improvement**:

* Experiment with more advanced feature engineering and selection techniques to better capture the underlying patterns in the data.
* Consider collecting more data, especially for underrepresented classes, to improve the model's learning capability.
* Hyperparameter tuning of the models might yield better results, as ensemble methods like Random Forest and Gradient Boosting have several parameters that can significantly influence performance.

#### **Conclusion**:

* The dataset presents a complex classification challenge, as evidenced by the varied performance across different models and techniques.
* The best-performing models are those that effectively handle class imbalance and leverage ensemble methods to deal with the dataset's complexity. Further experimentation and refinement are needed to improve performance, particularly for classes that currently have lower predictive accuracy.
